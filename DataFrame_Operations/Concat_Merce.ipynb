MERGE

```python
df1_merge= pd.read_csv('7-merge_data1.csv')
```

```python
df1_merge
```

```python
df2_merge=pd.read_csv('7-merge_data2.csv')
```

```python
df2_merge
```

```python
# merge - outer loin ==>veri biliminde farklı CSV dosyalarındaki tabloları ortak bir sütuna göre birleştirmek için kullanılır; outer join sayesinde iki tabloda da olmayan satırlar kaybolmaz, hepsi korunur.
```

```python
df_merge_outer= pd.merge(df1_merge,df2_merge, on="Employee_ID", how="outer")
```

```python
df_merge_outer
```

```python
# Merge - left join==>birleştirme yapılırken sol (ilk) tablodaki tüm satırların korunmasını, sağ tabloda eşleşmeyen yerlerin ise NaN kalmasını sağlar.
```

```python
df_merge_left= pd.merge(df1_merge,df2_merge,on="Employee_ID",how="left")
```

```python
df_merge_left
```

```python
# merge - rigt join ==> birleştirmede sağ (ikinci) tablodaki tüm satırların korunmasını, sol tabloda eşleşmeyen yerlerin ise NaN kalmasını sağlar.
```

```python
df_merge_right= pd.merge(df1_merge,df2_merge,on="Employee_ID",how="right")
```

```python
df_merge_right
```

```python
# merge - inner join ==>iki tablodaki sadece her iki tabloda da ortak olan satırların birleştirilmesini sağlar; eşleşmeyen satırlar kaybolur.
                        #NaN OLMAZ
```

```python
df_merge_inner=pd.merge(df1_merge,df2_merge,on="Employee_ID",how="inner")
```

```python
df_merge_inner
```
