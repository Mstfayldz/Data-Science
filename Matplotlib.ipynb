# Matplotlib

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#jupiter nootbokta grafik çizimini göstermesi  için gerekli
```

```python
age_list=[20,30,30,30,40,50,50,60,65,65]
weight_list=[50,50,65,70,70,75,80,80,85,85]
```

```python
plt.plot(age_list,weight_list,"r") # burda r =>red demek y,g,b farklı farklı kullanılabilir

#eksen isimlerini kendimiz belirleyebiliriz
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight by Age")

plt.show() # eğer notbook üzerinde çalışıyor olmasaydık sunucu üzerinde grafiği göstermeyebilirdi bu yüzden grafiği farklı kaydetmek durmundu kalırdık ==>plt.savefig("file_name") bu şekilde

```

Numpy ile

```python
np_age_list=np.array(age_list)
np_weight_list=np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight by Age")
plt.show()
```

```python
np_array1=np.linspace(0,10,20) #linspace burda 010 arasını eşit parçalara bölüp 20 tane değer yazıyor değerler arası farklar eşit
```

```python
np_array1
```

```python
np_array2=np_array1 ** 3 # üstel ==> exponential
```

```python
np_array2
```

```python
plt.plot(np_array1,np_array2 ,"b*-") # b* data pointleri de gösterir başka versionlar da vardır b* b-- gibi tabi bu blue için diğer renkler de olur
plt.title("Exponential-Graph")
plt.show()
```

Tek satırda birden fazla grafik gösterme

```python
plt.subplot(1,2,1)# burdaki sayısal değerler sırasıyla 1 row 2 column ve 1-> ilk graph
plt.plot(np_array1,np_array2 ,"b*-")
plt.subplot(1,2,2)# burdaki sayısal değerler sırasıyla 1 row 2 column ve 2-> ikinci graph
plt.plot(np_array2,np_array1 ,"r*-")
plt.show()
```

```python
# kendi figürümüzü oluşturmak ==>Kendi figürünü oluşturmak, veriyi başkalarının hazır şablonlarıyla değil soruna en uygun şekilde görselleştirmeni sağlar ve veriden çıkardığın içgörüyü daha açık, etkili ve ikna edici biçimde sunmana yardımcı olur. Böylece hem analizin güvenilirliği artar hem de hedef kitlen veriyi kolayca anlayabilir.
my_figure=plt.figure()
figure_axis=my_figure.add_axes([0.1,0.1,0.4,0.4]) # burdaki değerler sırasıyla
#1->left → figürün sol kenarından itibaren eksenin başlangıç konumu
#2->bottom → figürün alt kenarından itibaren eksenin başlangıç konumu
#3->width → eksenin genişliği
#4->height → eksenin yüksekliği
figure_axis.plot(np_array1,np_array2 ,"b")
figure_axis.set_xlabel("Age")
figure_axis.set_ylabel("Weight")
figure_axis.set_title("Graph Title")
plt.show()
```

```python
new_figure=plt.figure(dpi=150) # dpi ==> çözünürlülük
new_figure_axis=new_figure.add_axes([0.1,0.1,0.6,0.6])
new_figure_axis.plot(np_array1,np_array1**2 , label="np_array1**2")
new_figure_axis.plot(np_array1,np_array1**3 , label="np_array1**3")
new_figure_axis.legend() # label ları belirtir    new_figure_axis.legend(loc=1) loc burda konum belirtir
plt.savefig("figure.png")
plt.show()
```

# Matplotlib Styles

```python
data1=np.linspace(0,10,20)
data2=data1**2
```

```python
 #plt.subplots()  # ==> bu bize aslında bir figüre döndürür figür ve axes olarak
```

```python
my_fig,my_axes=plt.subplots() # dediğimizde bize hazır figür almayı sağlar
my_axes.plot(data1,data2,"k" ,alpha=0.9) # alpha tonu ayarlar
my_axes.plot(data2,data1, color="green") # color=#24525646 olabilir
plt.show()
```

```python
(new_fig ,new_axes)=plt.subplots(dpi=100)
new_axes.plot(data1,data1+2, color="#1E0703",alpha=0.8,linestyle="-", linewidth=1.5)
new_axes.plot(data1,data1+3, color="#32CD32",alpha=0.8,linestyle="--", linewidth=2.8)
new_axes.plot(data1,data1+4, color="yellow",alpha=1,linestyle="-.", linewidth=3.8)
new_axes.plot(data1,data1+5, color="red",alpha=1,linestyle=":", linewidth=4.8)
new_axes.plot(data1,data1+6, color="#000000",marker="o",markersize=6,markerfacecolor="blue", alpha=0.8,linestyle="-", linewidth=4.8) # marker ile datapointleri belirtiyoruz
new_axes.plot(data1,data1+7, color="blue",marker="*",markersize=6, alpha=0.8,linestyle="-", linewidth=1.8)
plt.show()
```
