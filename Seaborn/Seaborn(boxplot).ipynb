```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Median,quartile(%25 %50 %75) , min max 1.5 *IQR, outlier

```python
data= np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
plt.figure(figsize=(6,4))
sns.boxplot(y=data) # plt.boxplot(data)
plt.ylabel("Data Value")             # Sonuç olarak tabloda mavı kutunun ortasındaki çizgi==>MEDYAN////  mavi kutunun kendisi dağılımın yoğun olduğu yer
plt.grid(True)# grid ızgara sağlıyor # alt ve üst çizgiler normal dağılımdaki sınırlar///  üstteki yuvarlar ikse outler(çıkıntı değer)
plt.show()
```

# Titanic examination

```python
df= sns.load_dataset('titanic')
```

```python
df
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.boxplot(x='class',y='age', data=df)
plt.title("Class and Age")
plt.xlabel("Class")
plt.ylabel("Age")
plt.subplot(1,2,2)
sns.boxplot(x='class', y='fare', data=df)
plt.title("Class and Fare")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout()# bu tabloyu daha güzel gösterir
plt.show()
```

```python
sex=df.groupby("who")
sex.count()
```

```python
sns.set_style("darkgrid")
plt.scatter(x="fare",y="age", data=df)
plt.title("Who and Age")
plt.xlabel("Who")
plt.ylabel("Age")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="fare",y="age", hue="who", data=df)
plt.title("Who and Age")
plt.xlabel("Who")
plt.ylabel("Age")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="fare",y="age", hue="who", style="who", data=df)
plt.title("Who and Age")
plt.xlabel("Who")
plt.ylabel("Age")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.displot(x='age' , hue="who", data=df, kind="kde")
plt.title("Who and Age")
plt.xlabel("Who")
plt.ylabel("Age")
plt.show()
```
