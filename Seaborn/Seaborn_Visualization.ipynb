# Variations of the Seaborn Graph

```python
plt.scatter(df['Height'],df['Weight']) #plt.scatter("Height", "Weight", data=df)  şeklinde de olur
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

seaborn themes sitesinden bakılabilir
 [Seaborn Themes](https://python-graph-gallery.com/104-seaborn-themes/)





```python
sns.set_style("darkgrid")#white
sns.scatterplot(x='Height',y='Weight',data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex",data=df) # !!!! hue ==> spesifik bir column kullanmak için
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
df["Medal"].unique()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='Height',y='Weight', hue="Medal",data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
df["Medal"][df["Medal"]=="Gold"].count()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex",style="Medal", data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex",style="Medal", size="Age",data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='Height',y='Weight',data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='Height',y='Weight',hue="Sex", data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
sns.set_style("white")
sns.displot(x='Height' , hue="Sex", data=df) # displot ==> distribution ( dağılım)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```

```python
sns.set_style("white")
sns.displot(x='Height' , hue="Sex", data=df, kind="kde") # displot ==> distribution ( dağılım)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```

```python
sns.barplot(x="Medal", y="Height",hue="Sex", data=df)
plt.xlabel('Medal')
plt.ylabel('Height')
plt.title('Height vs Medal')
plt.show()
```

```python
df["Season"].unique()
```

```python
sns.catplot(x="Medal" , y="Height", hue="Sex", col="Season", data=df ) # catplot => categorize  kind="bar" eklenebilir mesela
plt.xlabel('Medal')
plt.ylabel('Height')
plt.title('Height vs Medal')
plt.show()
```
